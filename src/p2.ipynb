```python
# here we are going to use our st4 embeddings: embeddings_st4 and model_st4
from sentence_transformers import SentenceTransformer, util
model_st4 = SentenceTransformer('all-mpnet-base-v2')
query = "updated the security mirror for the Debian repositories"
query_vec = model_st4.encode([query])[0]
sen=[]
similar=[]

    
```

```python
my_sentences = []
with open('/content/data.txt', 'r', encoding='utf-8', errors='ignore') as filehandle:
    my_sentences = [elements.rstrip().replace('.', '') for elements in filehandle.readlines()]
```

```python
for s in my_sentences:
  sim = util.pytorch_cos_sim(query_vec, model_st4.encode([s])[0])
  sen.append(s)
  similar.append(sim)
  #   print("Sentence = ", s, "; similarity = ", sim)
```

```python
import pandas as pd
df_p2 = pd.DataFrame()
df_p2['Sentences'] = sen
df_p2['Similarity'] = similar
```

```python
lst = []
for s in similar:
  lst.append(s[0][0].item())
```

```python
lst
```

```python
df_p2['Similarity'] = lst
df_p2.head()
```

```python
df_p2.to_csv("p2.csv", encoding='utf-8')
```
